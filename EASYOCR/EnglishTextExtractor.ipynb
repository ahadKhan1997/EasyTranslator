```python
import easyocr
```

```python
#Download the model
reader = easyocr.Reader(['en'], gpu=False)
```

```python
import PIL
from PIL import ImageDraw
img = PIL.Image.open("wonderfulWorld.jpg")
img
```

```python
# Get bounding boxes
bounds = reader.readtext("wonderfulWorld.jpg")
bounds
```

```python
#Drawing bounding boxes
def draw_boxes(image, bounds, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(img, bounds)
```
